```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pickle
```

```python
with open('result.pkl', 'rb') as f:
    result = pickle.load(f)
```

```python
df = pd.DataFrame(result)
```

```python
df
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(x = df.columns, y = df.values[0], label="Single Thread", color="red")
ax.scatter(x = df.columns, y = df.values[1], label="Multi Thread", color="blue")
ax.legend()
plt.xlabel("Matrix Size")
plt.ylabel("Time (s)")
plt.title("Matrix Multiplication Time")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(df.columns, df.values[0], label="Single Thread", color="red")
ax.plot(df.columns, df.values[1], label="Multi Thread", color="blue")
ax.legend()
plt.xlabel("Matrix Size")
plt.ylabel("Time (s)")
plt.title("Matrix Multiplication Time")
plt.show()
```
